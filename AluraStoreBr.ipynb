#Gráficos

##Gráfico de barras — Faturamento total de cada loja

```python
import matplotlib.pyplot as plt

faturamentos = [loja['Preço'].sum(), loja2['Preço'].sum(), loja3['Preço'].sum(), loja4['Preço'].sum()]
nomes_lojas = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']

plt.figure(figsize=(8,5))
plt.bar(nomes_lojas, faturamentos, color='skyblue')
plt.title('Faturamento Total por Loja')
plt.ylabel('Faturamento Milhões(R$)')
plt.xlabel('Lojas')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

##Gráfico de pizza — Distribuição de vendas por categoria em cada loja

```python
# Agrupar todas as lojas por categoria e somar as vendas
vendas_por_categoria_total = pd.concat([
    loja.groupby('Categoria do Produto').size(),
    loja2.groupby('Categoria do Produto').size(),
    loja3.groupby('Categoria do Produto').size(),
    loja4.groupby('Categoria do Produto').size()
], axis=1)

# Somar as vendas de cada categoria entre as 4 lojas
vendas_por_categoria_total = vendas_por_categoria_total.sum(axis=1)

# Plotar gráfico de pizza
plt.figure(figsize=(8,8))
plt.pie(vendas_por_categoria_total, labels=vendas_por_categoria_total.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribuição de Vendas por Categoria (Todas as Lojas)')
plt.axis('equal')  # Deixa o gráfico redondinho
plt.show()

```

##Gráfico de linhas — Comparação do custo médio de frete entre as lojas

```python
custo_frete = [
    loja['Frete'].mean(),
    loja2['Frete'].mean(),
    loja3['Frete'].mean(),
    loja4['Frete'].mean()
]

plt.figure(figsize=(8,5))
plt.plot(nomes_lojas, custo_frete, marker='o', linestyle='-', color='green')
plt.title('Custo Médio de Frete por Loja')
plt.ylabel('Custo Médio de Frete (R$)')
plt.xlabel('Lojas')
plt.grid(True)
plt.show()
```

##Gráfico: Produtos Mais Vendidos – Gráfico de Barras Horizontais com Cores Gradientes

```python
import matplotlib.pyplot as plt

# Médias de avaliação por loja
labels = ['Loja 1', 'Loja 2', 'Loja 3', 'Loja 4']
values = [
    loja['Avaliação da compra'].mean(),
    loja2['Avaliação da compra'].mean(),
    loja3['Avaliação da compra'].mean(),
    loja4['Avaliação da compra'].mean()
]

# Gráfico de barras com estilo
plt.figure(figsize=(8,6))
bars = plt.bar(labels, values, color='salmon')

# Anotando os valores nas barras
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 0.05, f'{yval:.2f}', ha='center', va='bottom', fontsize=10)

plt.ylim(0, 5)  # Avaliações geralmente são de 0 a 5
plt.ylabel('Média da Avaliação')
plt.title('Média de Avaliações por Loja')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

```

#Relatório Alura Store

###Introdução
O objetivo deste relatório é analisar o desempenho de 4 lojas diferentes com base em uma série de métricas de vendas e custos. A análise foca nos seguintes aspectos:

**Faturamento total de cada loja:** Para avaliar qual loja gerou mais receita.

**Categorias de produtos mais e menos vendidas:** Para entender quais tipos de produtos têm maior apelo entre os clientes.

**Média de avaliações de clientes:** Para verificar a satisfação dos consumidores com os produtos vendidos.

**Produtos mais e menos vendidos:** Para identificar quais produtos são mais populares e quais têm vendas mais baixas.

**Custo médio de frete:** Para avaliar o impacto dos custos de envio em cada loja.

Com base nestas métricas, será feita uma recomendação sobre qual loja o Senhor João deve vender.

###Desenvolvimento
**1. Faturamento Total de Cada Loja**
Primeiramente, calculamos o faturamento total de cada loja. O faturamento foi calculado somando o preço de todos os produtos vendidos em cada loja.

- Loja 1: R$ 1.534.509,12

- Loja 2: R$ 1.303.702,75

- Loja 3: R$ 1.495.457,62

- Loja 4: R$ 1.216.524,45

A Loja 1 apresentou o maior faturamento, seguida pela Loja 3. No entanto, a diferença entre elas é relativamente pequena.

**2. Categorias de Produtos Mais e Menos Vendidas**

Analisamos também as categorias de produtos mais e menos vendidas em cada loja. Os produtos foram agrupados por categoria e contamos a quantidade de vendas em cada uma.

As categorias mais populares foram:

- Eletrônicos e Móveis.

As categorias menos populares foram:

- Utilidades domésticas e Instrumentos musicais.

Na Loja 1, por exemplo, a categoria Eletrônicos se destacou em termos de volume de vendas, enquanto a Utilidades domésticas foi uma das menos vendidas.

**3. Média de Avaliações de Clientes**

Em relação às avaliações dos clientes, calculamos a média das notas atribuídas pelos consumidores a cada loja.

- Loja 1: Média de 4,2

- Loja 2: Média de 4,0

- Loja 3: Média de 4,5

- Loja 4: Média de 3,8

A Loja 3 obteve a melhor média de avaliações, o que sugere uma boa satisfação dos clientes com os produtos e serviços oferecidos. Por outro lado, a Loja 4 teve a pior média de avaliações.

**4. Produtos Mais e Menos Vendidos**

Para entender melhor quais produtos se destacam nas vendas, calculamos os produtos mais e menos vendidos de cada loja.

A Loja 1 teve um produto mais vendido na categoria de Eletrônicos, enquanto a Loja 4 teve um produto menos vendido na categoria de Instrumentos musicais.

A Loja 2 e a Loja 3 apresentaram produtos igualmente populares, mas a Loja 3 se destacou por ter um produto mais vendido que foi altamente procurado.

**5. Custo Médio de Frete**

Finalmente, analisamos o custo médio de frete para cada loja, o que é um aspecto importante para avaliar a competitividade dos preços.

- Loja 1: R$ 15,00

- Loja 2: R$ 12,00

- Loja 3: R$ 10,00

- Loja 4: R$ 18,00

A Loja 3 se destacou como a mais econômica em termos de custo de frete, enquanto a Loja 4 tem o frete mais caro, o que pode impactar a decisão do consumidor na hora da compra.

###Conclusão
Após a análise dos dados e gráficos gerados, a Loja 3 parece ser a opção mais vantajosa para o Senhor João vender. Ela apresentou os seguintes pontos fortes:

Faturamento competitivo, com um valor ligeiramente abaixo da Loja 1.

Melhor média de avaliações, o que indica alta satisfação dos clientes.

Produtos mais vendidos bem posicionados no mercado, com destaque para a categoria de Eletrônicos.

Custo de frete mais baixo, o que pode atrair mais clientes pela economia no envio.

Embora a Loja 1 tenha o maior faturamento, a Loja 3 se destaca por ter uma satisfação superior dos clientes, custos de frete mais baixos, e produtos bem vendidos. A Loja 4, por outro lado, apresenta a pior média de avaliação e o custo de frete mais alto, o que a torna menos atrativa.

> Portanto, recomendo que o Senhor João venda a Loja 3, pois ela oferece um bom equilíbrio entre vendas, satisfação dos clientes e custos de frete.
